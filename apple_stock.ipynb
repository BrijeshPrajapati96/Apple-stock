```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
stock= pd.read_csv('Downloads\AppleStockPrice.csv')
stock.head()
```

```python
stock.info()
```

```python
stock['Date'] = pd.to_datetime(stock['Date'], dayfirst=True)
stock.set_index('Date', inplace=True)
```

```python
#plotting the closing tab
plt.figure(figsize=(15, 10))
stock['Adj Close'].plot()
plt.ylabel('Close Price')
plt.xlabel('Date')
plt.title("Closing Price of APL")
plt.tight_layout()
plt.show()
```

```python
# 2. Plotting the Sales Volume
plt.figure(figsize=(15, 10))
stock['Volume'].plot()
plt.ylabel('Volume')
plt.xlabel('Date')
plt.title("Sales Volume of AAPL")
plt.tight_layout()
plt.show()
```

```python
# 3. Calculate Moving Averages (10, 20, 50 days)
ma_days = [10, 20, 50]
for ma in ma_days:
    stock[f"MA for {ma} days"] = stock['Adj Close'].rolling(ma).mean()
    
# Plot the Moving Averages with Adjusted Close Price
plt.figure()
stock[['Adj Close', 'MA for 10 days', 'MA for 20 days', 'MA for 50 days']].plot()
plt.title('APPLE Adjusted Close Price and Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.tight_layout()
plt.show()
```

```python
# 4. Calculate Daily Return
stock['Daily Return'] = stock['Adj Close'].pct_change()

# Plot the Daily Return
plt.figure(figsize=(15, 10))
stock['Daily Return'].plot(legend=True, linestyle='--', marker='o')
plt.title('APPLE Daily Return')
plt.xlabel('Date')
plt.ylabel('Daily Return')
plt.tight_layout()
plt.show()
```

```python
# 5. Histogram of Daily Returns
plt.figure(figsize=(12, 9))
stock['Daily Return'].hist(bins=50)
plt.xlabel('Daily Return')
plt.ylabel('Counts')
plt.title('Apple Daily Return Distribution')
plt.tight_layout()
plt.show()
```
